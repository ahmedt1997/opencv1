```python
import matplotlib.pyplot as plt
import cv2
import matplotlib.image as mpimg
import numpy as np
```

```python
#READ AND DISPLAY IMAGE
image_color = mpimg.imread('auto_route.jpg')
```

```python
image_color.shape
```

```python
plt.imshow(image_color)
```

```python
#convert this image to greyscale
```

```python
image_grey=cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
```

```python
image_grey.shape
```

```python
plt.imshow(image_grey,cmap='gray')
```

```python
#save the image in greyscale
```

```python
cv2.imwrite('auto_route_greyscale.jpg',image_grey)
```
